```python
import pandas as pd
import os
from pathlib import Path
from google.colab import drive

# 1. Montar Google Drive
drive.mount('/content/drive')

# 2. Ruta del archivo TSV en Drive
# Cambiá esta ruta según dónde tengas guardado el archivo
archivo = "/content/drive/MyDrive/farmacos/Farmaco/TP FINAL/s001A.dali.tsv"

with open(archivo, "r", encoding="utf-8", errors="ignore") as f:
    lineas = f.readlines()

inicio_tabla = None

for i, linea in enumerate(lineas):
    # Una línea TSV real debería tener varias tabulaciones
    if linea.count("\t") >= 5:
        inicio_tabla = i
        break

if inicio_tabla is None:
    raise ValueError("No pude encontrar una tabla separada por tabulaciones en el archivo.")

print("La tabla parece empezar en la línea:", inicio_tabla + 1)

# Leer desde esa línea
df = pd.read_csv(
    archivo,
    sep="\t",
    skiprows=inicio_tabla,
    encoding="utf-8",
    engine="python"
)

df.head()
```

```python
# Limpiar nombres de columnas
df.columns = (
    df.columns
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace(".", "", regex=False)
)

print(df.columns)
df.head()
```

```python
# Pasar a numérico
df["z_score"] = pd.to_numeric(df["z_score"], errors="coerce")
df["rmsd"] = pd.to_numeric(df["rmsd"], errors="coerce")
df["ali_length"] = pd.to_numeric(df["ali_length"], errors="coerce")

# Filtro estructural pedido por la consigna
df_filtrado = df[
    (df["z_score"] > 6) &
    (df["rmsd"] < 3.5) &
    (df["ali_length"] >= 50)
].copy()

# Filtro preliminar de humanos por descripción
"""df_filtrado["posible_humano"] = df_filtrado[desc_col].astype(str).str.contains(
    "human|homo sapiens",
    case=False,
    na=False
)"""

df_filtrado = df_filtrado.sort_values(
    by=["z_score", "ali_length"],
    ascending=[False, False]
)

df_filtrado.head(30)
```

```python
salida = "/content/drive/MyDrive/farmacos/Farmaco/TP FINAL/dali_candidatos_filtrados.csv"

df_filtrado.to_csv(salida, index=False)

print("Archivo guardado en:", salida)
```
